```python
import matplotlib.pyplot as plt
from simulator import RaceSimulator

sim = RaceSimulator(20, "soft")

lap_times = []
laps = []


while True:
    if sim.current_lap == 8:
        action = 3  # pit to hard
    else:
        action = 0

    state, reward, done = sim.step(action)

    lap_times.append(-reward)
    laps.append(state[0] - 1)

    if done:
        break

plt.plot(laps, lap_times, marker='o')
plt.xlabel("Lap")
plt.ylabel("Lap Time")
plt.title("F1 Race Simulation")
plt.show()
```

```python
import matplotlib.pyplot as plt
from simulator import RaceSimulator
import time

sim = RaceSimulator(20, "soft")

laps = []
lap_times = []

plt.ion()

while True:
    if sim.current_lap == 8:
        action = 3
    else:
        action = 0

    state, reward, done = sim.step(action)

    laps.append(state[0] - 1)
    lap_times.append(-reward)

    plt.clf()
    plt.plot(laps, lap_times, marker='o')
    plt.xlabel("Lap")
    plt.ylabel("Lap Time")
    plt.title("Live Race Simulation")

    plt.pause(0.5)

    if done:
        break

plt.ioff()
plt.show()
```
